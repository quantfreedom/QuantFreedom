```python
import pandas as pd
import numpy as np

from quantfreedom.backtester.indicators.talib_ind import from_talib
from quantfreedom.backtester.evaluators.evaluators import eval_is_below, eval_is_above
prices = pd.read_csv('30min.csv', index_col='time')
```

```python
rsi_ind = from_talib(
    func_name='rsi',
    df_prices=prices,
    cart_product=False,
    combos=False,
    timeperiod=[15, 20, 30],
)
```

```python
eval_is_below(
    rsi_ind,
    np.arange(40, 61, 5)
)
```

```python
eval_is_above(
    rsi_ind,
    np.arange(60, 39, -5)
)
```
